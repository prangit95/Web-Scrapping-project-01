Import required libraries

```python
from bs4 import BeautifulSoup
import requests
```

url for web scrapping

```python
url = 'http://www.simplilearn.com'
```

```python
#access the url with request obj
result = requests.get(url)
#scrape the web page content
webpage = result.content
#create a soup obj for parsing web page using html parser
sl_soup = BeautifulSoup(webpage, 'html.parser')
```

```python
#close the result obj
result.close()
```

View the content of the soup object

```python
sl_soup.contents
```

```python
#using pretify method to view well formated output
print(sl_soup.prettify())
```

```python
#view the head of the soup obj
print(sl_soup.head)
```

```python
#view the title of the web page
print(sl_soup.title)
```

```python
#find the all links present in the web page
for href in sl_soup.findAll('a', href=True):
    print(href['href'])
```
